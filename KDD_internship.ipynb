```python
import re

def preprocess(text):
    text = re.sub(r'\s+', ' ', text)  # Remove multiple spaces
    text = re.sub(r'\[[^]]*\]', '', text)  # Remove references
    text = re.sub(r'\([^)]*\)', '', text)  # Remove parentheses
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = text.lower()
    return text

data = data.map(lambda x: {"article": preprocess(x["article"]), "abstract": preprocess(x["abstract"])})
```

```python
from transformers import pipeline, AutoModelForSeq2SeqLM, AutoTokenizer
import ipywidgets as widgets
from IPython.display import display
from datasets import load_dataset
import re

# Load the dataset
dataset = load_dataset("ccdv/pubmed-summarization")

# Initialize the model and tokenizer
model_name = "sshleifer/distilbart-cnn-12-6"
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Initialize the summarization pipeline
summarizer = pipeline("summarization", model=model, tokenizer=tokenizer)


# Summarization function
def summarize_article(article):
    try:
        summary = summarizer(article, max_length=100, min_length=30, do_sample=False)
        summary_text = summary[0]['summary_text']
    except Exception as e:
        summary_text = f"Error in summarization: {str(e)}"
    return summary_text

# Create widgets
upload_widget = widgets.FileUpload(
    accept='.txt',  # Accepted file types
    multiple=False  # Accept single file only
)

original_text_widget = widgets.Textarea(
    value='',
    placeholder='Original article will appear here...',
    description='Original:',
    disabled=False,
    layout=widgets.Layout(width='100%', height='200px')
)

summarized_text_widget = widgets.Textarea(
    value='',
    placeholder='Summary will appear here...',
    description='Summary:',
    disabled=False,
    layout=widgets.Layout(width='100%', height='200px')
)

# Display the widgets
display(upload_widget, original_text_widget, summarized_text_widget)

# Handle file upload and summarization
def on_upload_change(change):
    uploaded_file = upload_widget.value
    if uploaded_file:
        file_info = uploaded_file[list(uploaded_file.keys())[0]]
        article_text = file_info['content'].decode('utf-8')
        article_text = preprocess(article_text)
        original_text_widget.value = article_text
        summary = summarize_article(article_text)
        summarized_text_widget.value = summary

# Link the function to the file upload widget
upload_widget.observe(on_upload_change, names='value')
```

```python
from flask import Flask, render_template, request
from flask_ngrok import run_with_ngrok
from transformers import pipeline, AutoModelForSeq2SeqLM, AutoTokenizer
from datasets import load_dataset
import re

app = Flask(__name__)
run_with_ngrok(app)  # Start ngrok when app is run

# Load the PubMed summarization dataset
dataset = load_dataset("ccdv/pubmed-summarization")

# Initialize the model and tokenizer
model_name = "sshleifer/distilbart-cnn-12-6"
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Initialize the summarization pipeline
summarizer = pipeline("summarization", model=model, tokenizer=tokenizer)

@app.route('/', methods=['GET', 'POST'])
def index():
    if request.method == 'POST':
        if 'file' in request.files:
            file = request.files['file']
            if file.filename != '':
                article_text = file.read().decode("utf-8")
                return summarize(article_text)
        elif 'text' in request.form:
            article_text = request.form['text']
            return summarize(article_text)

        return render_template('index.html', error="Please upload a valid text file or input text.")
    else:
        return render_template('index.html')

def summarize(article_text):
    article_text = preprocess(article_text)
    try:
        summary = summarizer(article_text, max_length=100, min_length=30, do_sample=False)
        summarized_text = summary[0]['summary_text']
        return render_template('index.html', original_text=article_text, summarized_text=summarized_text)
    except Exception as e:
        return render_template('index.html', error=f"Error in summarization: {str(e)}")

if __name__ == '__main__':
    app.run()
```
