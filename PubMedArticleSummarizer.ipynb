```python
%%writefile app.py

import streamlit as st
from transformers import T5ForConditionalGeneration, T5Tokenizer
import re
from pymongo import MongoClient
from bcrypt import hashpw, gensalt, checkpw

# MongoDB setup (Atlas instance)
client = MongoClient("mongodb+srv://i211763:1234@cluster1.flqjs9y.mongodb.net/?retryWrites=true&w=majority")
db = client['summarizer']
users_collection = db['users']
summaries_collection = db['summaries']

# Function to hash passwords
def hash_password(password):
    return hashpw(password.encode('utf-8'), gensalt())

# Function to verify password
def verify_password(stored_password, provided_password):
    return checkpw(provided_password.encode('utf-8'), stored_password)

# Preprocess function
def preprocess(text):
    text = re.sub(r'\s+', ' ', text)  # Remove multiple spaces
    text = re.sub(r'\[[^]]*\]', '', text)  # Remove references
    text = re.sub(r'\([^)]*\)', '', text)  # Remove parentheses
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = text.lower()
    return text

# Initialize T5 models and tokenizers
model_names = {
    "brief": "t5-small",
    "detailed": "t5-base"
}
models = {style: T5ForConditionalGeneration.from_pretrained(name) for style, name in model_names.items()}
tokenizers = {style: T5Tokenizer.from_pretrained(name) for style, name in model_names.items()}

# Function to summarize an article
def summarize_article(article, style):
    article = preprocess(article)
    tokenizer = tokenizers[style]
    model = models[style]
    inputs = tokenizer.encode("summarize: " + article, return_tensors="pt", max_length=512, truncation=True)

    if style == "brief":
        max_length = 50
        min_length = 10
    else:
        max_length = 300
        min_length = 30

    summary_ids = model.generate(inputs, max_length=max_length, min_length=min_length, length_penalty=2.0, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

# Streamlit app with authentication
def main():
    st.title("PubMed Article Summarizer")

    if 'logged_in' not in st.session_state:
        st.session_state.logged_in = False

    if 'username' not in st.session_state:
        st.session_state.username = ""

    # Authentication
    menu = ["Login", "Register"]
    choice = st.sidebar.selectbox("Menu", menu)

    if choice == "Login":
        st.subheader("Login Section")
        username = st.sidebar.text_input("Username")
        password = st.sidebar.text_input("Password", type='password')
        if st.sidebar.button("Login"):
            user = users_collection.find_one({"username": username})
            if user and verify_password(user['password'], password):
                st.session_state.logged_in = True
                st.session_state.username = username
                st.success(f"Welcome {username}")
            else:
                st.error("Invalid username or password")

    elif choice == "Register":
        st.subheader("Register Section")
        new_username = st.sidebar.text_input("New Username")
        new_password = st.sidebar.text_input("New Password", type='password')
        if st.sidebar.button("Register"):
            if new_username and new_password:
                existing_user = users_collection.find_one({"username": new_username})
                if existing_user is None:
                    hashed_password = hash_password(new_password)
                    users_collection.insert_one({"username": new_username, "password": hashed_password})
                    st.sidebar.success("User registered successfully")
                else:
                    st.sidebar.error("Username already exists")

    if st.session_state.logged_in:
        st.subheader("Summarization Section")

        # Option to upload a file
        uploaded_file = st.file_uploader("Upload an article file (txt)", type="txt")
        article = ""
        if uploaded_file is not None:
            article = uploaded_file.read().decode("utf-8")
            st.text_area("Uploaded Article:", value=article, height=200, disabled=True)
            st.session_state.article = article
        else:
            article = st.text_area("Or enter the article to summarize:", value=st.session_state.get('article', ''))

        # Select summarization style
        summary_style = st.radio("Choose summary style:", ("brief", "detailed"))

        if st.button("Summarize"):
            if article.strip() == "":
                st.error("Please enter a valid article.")
            else:
                summary = summarize_article(article, summary_style)
                st.subheader("Original Article:")
                st.write(article)
                st.subheader("Summary:")
                st.write(summary)

                # Save the summary
                summaries_collection.insert_one({"username": st.session_state.username, "article": article, "summary": summary, "style": summary_style})
                st.success("Summary saved successfully.")

        st.header("Your Summaries")
        summaries = summaries_collection.find({"username": st.session_state.username})
        for summary in summaries:
            st.write(f"**Style**: {summary['style']}")
            st.write(f"**Original Article**: {summary['article']}")
            st.write(f"**Summary**: {summary['summary']}")
            st.write("---")

if __name__ == "__main__":
    main()


```
